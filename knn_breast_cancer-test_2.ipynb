```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt #matplotlib is used for plot the graphs,
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
df.drop("Unnamed: 32", axis = 1, inplace = True)
df.shape
df.head()
```

```python
df["diagnosis"]= df["diagnosis"].map({'B':0,'M':1}).astype(int)
df.head()
```

```python
x=df[['concave points_worst', 'perimeter_worst', 'concave points_mean', 'radius_worst', 'perimeter_mean', 'area_worst', 'radius_mean', 'area_mean', 'concavity_mean', 'concavity_worst', 'compactness_mean', 'compactness_worst', 'radius_se', 'perimeter_se', 'area_se']]
y=df[['diagnosis']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print ("Shape of Train Data:", x_train.shape)
print ("Shape of Test Data:", x_test.shape)
```

```python
print ("Shape of Train Data:", y_train.shape)
print ("Shape of Test Data:", y_test.shape)
```

```python
from sklearn.metrics import accuracy_score
model = KNeighborsClassifier(n_neighbors=3)
model.fit(x_train, y_train.values.ravel())
```

```python
predict = model.predict(x_test)
```

```python
x = pd.DataFrame(predict)
x.shape
```

```python
accuracy_score(predict,y_test)
```

```python

print(accuracy_score)
```

```python
model.score(x_train,y_train)
```

```python
predict.max()
```

```python
##model evaluation
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predict))
```

```python
cm =confusion_matrix(y_test, predict)
sns.heatmap(cm, square=True , annot=True)
```

```python
#Optimal value of K
error_rate = []
# Might take some time
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train.values.ravel())
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test.values.ravel()))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
